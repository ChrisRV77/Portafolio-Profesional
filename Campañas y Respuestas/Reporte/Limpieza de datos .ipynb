```python
#librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_marketing = pd.read_csv('marketing_raw.csv', sep=',')
df_marketing = df_marketing.replace({r'\$': ''}, regex=True)
df_marketing = df_marketing.astype(float)
df_marketing.head()
```

# Buscamos las colmunas que pueden ser categoricas por la cantidad distinta de valores que tengan en sus columnas

```python
var_columnas = df_marketing.columns
# Umbral: Si tiene menos de 10 valores únicos, la consideramos categórica
threshold = 10

categoricas_potenciales = []
numericas_reales = []

for col in df_marketing.select_dtypes(include=['number']).columns:
    cardinalidad = df_marketing[col].nunique()

    if cardinalidad <= threshold:
        categoricas_potenciales.append(col)
    else:
        numericas_reales.append(col)

print(f"Parecen ser categorías: {categoricas_potenciales}")
```

# Despivote de las columnas Marital_ y Education

```python
#melt a las columnas de marital_status
marital_status = ['marital_Divorced', 'marital_Married', 'marital_Single', 'marital_Together', 'marital_Widow']
var_no_marital = list(set(var_columnas).difference(marital_status))
df_marketing_melt = pd.melt(df_marketing, id_vars=var_no_marital, value_vars=marital_status)
df_marketing_melt = df_marketing_melt[df_marketing_melt['value'] == 1].drop('value', axis=1)
df_marketing_melt = df_marketing_melt.rename(columns={'variable': 'marital_status'})
df_marketing_melt['marital_status'] = df_marketing_melt['marital_status'].str.replace('marital_', '')

#melt a las columnas de education_status
var_columnas_melt = df_marketing_melt.columns
education_status = ['education_2n Cycle', 'education_Basic', 'education_Graduation', 'education_Master', 'education_PhD']
var_no_education = list(set(var_columnas_melt).difference(education_status))
df_marketing_melt = pd.melt(df_marketing_melt, id_vars=var_no_education, value_vars=education_status)
df_marketing_melt = df_marketing_melt[df_marketing_melt['value'] == 1].drop('value', axis=1)
df_marketing_melt = df_marketing_melt.rename(columns={'variable': 'education_status'})
df_marketing_melt['education_status'] = df_marketing_melt['education_status'].str.replace('education_', '')
```

#Hacemos un analisis de las variables para buscar correlaciones y anomalias en el dataset

R: No se encuentran anomalías

```python
non_relevant_or_categories_var = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response','AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmpOverall', 'marital_status', 'education_status', 'Kidhome', 'Teenhome', 'Complain', 'Z_Revenue', 'Z_CostContact' ]
df_filtrado = df_marketing_melt.loc[:, ~df_marketing_melt.columns.isin(non_relevant_or_categories_var)]

#grafico para hacer comparacion contra otra variable y buscar anomalias
sns.pairplot(df_filtrado, diag_kind='kde', plot_kws={'alpha': 0.6})

plt.show()
```

```python
var_columns_mnt = [i for i in var_columnas_melt if 'Mnt' in i]
sns.pairplot(df_filtrado[var_columns_mnt], diag_kind='kde', plot_kws={'alpha': 0.6})
plt.show()
```
